```python
class Celular:
    # Constructor
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio

    # Método para mostrar información
    def mostrar_info(self):
        return f"{self.marca} {self.modelo} - ${self.precio}"

    def encender(self):
        print(f"El celular {self.marca} {self.modelo} está encendido.")

# Instanciar objetos
celular1 = Celular("Samsung", "Galaxy S23", 800)
celular2 = Celular("Apple", "iPhone 15", 1000)

print("Celular1 :")
print(celular1.mostrar_info())
celular1.encender()

print("Celular2 :")
print(celular2.mostrar_info())
celular2.encender()

print("Programa terminado")
print("Creador de software Jesus Arriaga NL = 5")
```
